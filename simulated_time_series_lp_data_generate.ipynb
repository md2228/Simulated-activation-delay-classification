```python
import random
import pandas as pd
```

```python
# Generate simulated timing samples
amount_of_rows = 1000
li = []
# No delay samples first
for j in range((int)(amount_of_rows/2)):
    arr = []
    for i in range(5):
        arr.append(random.uniform(0, 1))

    arr.sort(key = float)
    li.append(arr)

# Samples with added delay
for j in range((int)(amount_of_rows/2)):
    arr = []
    for i in range(5):
        r1 = random.uniform(0, 1) * 0.1
        arr.append(random.uniform(0, 1) + (r1 * i))

    arr.sort(key = float)
    li.append(arr)
li
```

```python
# Create a dataframe and add the list items from the previous block

df = pd.DataFrame(columns =  ["Pin1", "Pin2", "Pin3", "Pin4", "Pin5", "Method"])

for i in range(amount_of_rows):
    temp_li = li[i]
    if i >= ((int)(amount_of_rows/2)):
        temp_li.append('Picked')
        df.loc[-1] = temp_li # adding a row
        
    else:
        temp_li.append('Key')
        df.loc[-1] = temp_li  # adding a row
        
    df.index = df.index + 1  # shifting index
    df = df.sort_index()

df
```

```python
# Save the created dataframe to a .csv
df.to_csv('simulated_time_series_lp_data_validation.csv')
```
